```python
data = pd.read_csv('data.csv')
```

```python
data.describe()
```

```python
for x in data.columns:
    if x.startswith('cat'):
        print(x, data[x].unique())
```

Notice cat99 doesn't start naming categories with "A", so its possible the meaning of levels for cat99 is shared with other columns

```python
for col in data.columns:
    if col.startswith('cat'):
        data[col]=data[col].astype('category')
```

```python
data.hist(figsize=(14,14))
```

```python
from pandas.tools.plotting import scatter_matrix
```

```python
scatter_matrix(data, alpha=0.2, figsize=(12, 12), diagonal='kde')
```

```python
for col in data.columns:
    if type(data[col].dtype) == pd.types.dtypes.CategoricalDtype:
        dummies=pd.get_dummies(data[col], prefix=col)
        data=pd.concat([data, dummies], axis=1)
        del data[col]
```

```python
data.head()
```

```python
y= data['loss']
X = data.drop(['loss', 'id'], axis = 1)
```

```python
X.columns
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)
```

```python
from sklearn.svm import SVR
```

```python
clf = SVR(C=1.0, epsilon=0.2)
```

```python
clf.fit(X_train, y_train) 
```

```python
yhat=clf.predict(X_test)
```

Coefficient of determination R^2

```python
clf.score(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators = 100)
```

```python
rf.fit(X_train, y_train)
```

```python
yhat_rf=rf.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_test, yhat_rf)
```

```python
import xgboost as xgb
```

```python
xg_train = xgb.DMatrix(X_train, label=y_train)
xg_test = xgb.DMatrix(X_test, label=y_test)
```

```python
#TRAIN
# setup parameters for xgboost
param = {}
param['objective'] = 'reg:linear'
#param['eta'] = 1
#param['gamma']= 4
#param['max_depth'] = 8
#param['min_child_weight']=4
#param['max_delta_step']=4
#param['subsample']=0.5

param['nthread'] = 3
#param['subsample']=1
param['num_class'] = 3
param['eval_metric']='rmse'


watchlist = [(xg_train,'train'), (xg_test, 'test')]
num_round = 6
bst = xgb.train(param, xg_train, num_round, evals=watchlist)

print(xgb.cv(param, xg_train, num_round, nfold=3, seed=0))
```
